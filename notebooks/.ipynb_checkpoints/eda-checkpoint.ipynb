# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

# Exploratory Data Analysis (EDA)

#### 1. Read the Data CSV

```python
df = pd.read_csv("../data/raw/job_postings_processed.csv")
df.head()
```

#### 2. Get a Summary of the Data

```python
print("DataFrame Shape:", df.shape)
```

```python
df.info()
```

```python
df.describe()
```

#### 3. Check for Missing Values

```python
missing_percentage = df.isnull().sum() / len(df) * 100
print(missing_percentage)
```

#### 4. Column Distribution Visualization

```python
df.hist(bins=50, figsize=(20,15))
plt.show()
```

#### 5. Correlation Analysis

```python
numeric_df = df.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="crest")
plt.show()
```

#### 6. Analyzing Job Titles

```python
word_counts = Counter(" ".join(df["title"]).split())
most_common_words = word_counts.most_common(10)
for most_common_word in most_common_words:
    print(most_common_word)
```

#### 7. Salary Analysis

```python
df[['min_salary', 'max_salary']].plot(kind='box', vert=False, figsize=(10, 6))
plt.show()
```

#### 8. Work Location Analysis

```python
location_counts = df['location'].value_counts().head(10)
location_counts.plot(kind='barh')
plt.title('Top 10 Job Locations')
plt.show()
```

#### 9. Analyzing Application Types and Views

```python
application_type_counts = df['application_type'].value_counts()
application_type_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Application Types Distribution')
plt.show()
```
